# Plot neutron cross section

The data are from ENDF and collected from the IAEA website.

I downloaded the plotted (or interpolated or tabulated) data, at 293K. 
The data in endf file are at 0K and does not give the desired result.

```python
import csv
import matplotlib.pyplot as plt
```

```python
def Read_Two_Column_File(file_name):
    with open(file_name, 'r') as f_input:
        csv_input = csv.reader(f_input, delimiter=' ', skipinitialspace=True)
        x = []
        y = []
        for cols in csv_input:
            if cols[0].startswith('#') or cols[0].startswith('/'):
                continue
            x.append(float(cols[0]))
            y.append(float(cols[1]))

    return x, y
```

```python
plt.rcParams.update({'font.size': 14}) #14
plt.rcParams.update({'font.family': 'Latin Modern Roman'}) #serif
plt.rcParams.update({'font.weight': 'bold'})

font = {'family': 'Latin Modern Roman',
        'weight': 'bold',
        'size': 14,
        }
```

```python
u235Sx, u235Sy = Read_Two_Column_File('data/U-235(N,EL).txt')
u235Fx, u235Fy = Read_Two_Column_File('data/U-235(N,F).txt')
u235Cx, u235Cy = Read_Two_Column_File('data/U-235(N,G).txt')
#print(u235Sy)
```

```python
plt.plot(u235Fx, u235Fy, label='Fission',    color='blue', alpha=1, linewidth=1)
plt.plot(u235Cx, u235Cy, label='Capture',    color='red', alpha=0.8, linewidth=1)
plt.plot(u235Sx, u235Sy, label='Scattering', color='black', alpha=1, linewidth=1)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Incident energy (MeV)', fontdict=font)
plt.ylabel('Cross section (barns)', fontdict=font)
plt.legend(loc='lower left')
plt.title('Cross Sections for U235 at 293K', fontdict=font)
plt.tight_layout()
plt.savefig("U-235.pdf")
plt.savefig("png/U-235.png")
plt.show()
```

```python
u238Sx, u238Sy = Read_Two_Column_File('data/U-238(N,EL).txt')
u238Fx, u238Fy = Read_Two_Column_File('data/U-238(N,F).txt')
u238Cx, u238Cy = Read_Two_Column_File('data/U-238(N,G).txt')
```
