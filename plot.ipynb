```python
import csv
import matplotlib.pyplot as plt

def Read_Two_Column_File(file_name):
    with open(file_name, 'r') as f_input:
        csv_input = csv.reader(f_input, delimiter=' ', skipinitialspace=True)
        x = []
        y = []
        for cols in csv_input:
            if cols[0].startswith('#') or cols[0].startswith('/'):
                continue
            x.append(float(cols[0]))
            y.append(float(cols[1]))

    return x, y

```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rcParams.update({'font.size': 12})
```

```python
u235Sx, u235Sy = Read_Two_Column_File('U-235(N,EL).txt')
u235Fx, u235Fy = Read_Two_Column_File('U-235(N,F).txt')
u235Cx, u235Cy = Read_Two_Column_File('U-235(N,G).txt')
#print(x)
#print(y)
```

```python
plt.plot(u235Sx, u235Sy, label='Scattering', color='black', alpha=1, linewidth=0.45) #, linewidth=0.75
plt.plot(u235Fx, u235Fy, label='Fission',    color='blue', alpha=0.95, linewidth=0.45)
plt.plot(u235Cx, u235Cy, label='Capture',    color='red', alpha=0.9, linewidth=0.45)
plt.xscale('log')
plt.yscale('log')
#plt.ylim([0, 1])
plt.xlabel('Incident energy (MeV)')
plt.ylabel('Cross section (barns)')
plt.legend(loc='lower left')
#plt.title('Cross Sections for U235 at K')
plt.tight_layout()
plt.savefig("U-235.pdf")
plt.show()
```
